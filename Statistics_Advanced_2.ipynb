1. What is hypothesis testing in statistics?

* Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data.
* It helps us test assumptions (called hypotheses) about population parameters (like mean, proportion, variance).
* Key Steps in Hypothesis Testing
* **State the hypotheses:**
* Null hypothesis (H₀): The default assumption — usually states there is no effect or no difference.
Example: H0:μ=50 (population mean is 50).
* Alternative hypothesis (H₁ or Ha): What we want to test — usually states there is an effect or a difference.
Example: Ha:μ=50.
* **Choose significance level (α):**
* The probability of rejecting H0 when it is actually true (Type I error).
* Common values: 0.05 (5%), 0.01 (1%).
* **Select the test statistic:**
* Based on data type and sample size, e.g., z-test, t-test, chi-square test, ANOVA.
* **Calculate the test statistic and p-value:**
* Test statistic: A standardized value to compare against critical values.
* p-value: Probability of observing the data (or something more extreme) if H0 were true.
* **Decision rule:**
* If p-value ≤ α, reject H0→ evidence supports Ha.
* If p-value > α, fail to reject H0→ not enough evidence against H0.

2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?

* **Null Hypothesis (H₀)**
* The starting assumption in hypothesis testing.
* It usually states that there is no effect, no difference, or that a parameter equals a specific value.
* It represents the "status quo" or the claim we test against.
* Example:
   * A company says their battery lasts 100 hours.
   * Null hypothesis: H0:μ=100.(The average battery life is 100 hours.)
***Alternative Hypothesis (H₁ or Ha)**
* The claim we want to find evidence for.
* It states that there is an effect or a difference.
* It is accepted if the data provide strong enough evidence against H0.
* Example:
   * If we suspect the battery life is different from 100 hours:
   * Alternative hypothesis: Ha:μ=100.
   * If we suspect the battery life is less than 100 hours:Ha:μ<100.
   * If we suspect it’s more than 100 hours:Ha:μ>100.

3. Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test?

* The significance level is the threshold we set to decide whether to reject the null hypothesis.
* It represents the probability of making a Type I error (rejecting H0 when H0
	is actually true).
* Common choices:
  * 0.05 (5%) → most widely used.
  * 0.01 (1%) → stricter, used in medical/critical studies.
  * 0.10 (10%) → more lenient, sometimes used in exploratory research.
***Role in Decision Making**
* When we run a test, we calculate a p-value (the probability of observing the data, or something more extreme, if H0 were true).
* If p-value ≤ α → reject H0 (enough evidence against null).
* If p-value > α → fail to reject H0 (not enough evidence to support alternative).
* Example
   * A company claims the mean weight of its chips packets is 50g.We test with α = 0.05.
   * Null hypothesis: H0:μ=50
   * Alternative hypothesis: Ha:μ=50
   * Suppose we get p-value = 0.03.
   * Since 0.03 < 0.05, we reject H0.
* This means there’s enough evidence to say the mean weight is not 50g.

4.What are Type I and Type II errors? Give examples of each?

* **Type I Error (False Positive)**
* Rejecting the null hypothesis (H0) when it is actually true.
* It’s like saying “there is an effect” when in reality, there isn’t.
* Its probability = α (significance level).
* Example:
  * Null hypothesis (H0): A man is innocent.
  * We wrongly convict him (reject H0) even though he is actually innocent.
  *That’s a Type I error.
* Statistical example:
* H0: A medicine has no effect.
* We conclude the medicine works (reject H0) even though it doesn’t.
***Type II Error (False Negative)**
*  Failing to reject the null hypothesis (H0) when it is actually false.
* It’s like saying “no effect” when there really is one.
* Its probability = β. (And Power of the test = 1 − β).
* Example:
   * Null hypothesis (H0): A man is innocent.
   * He is actually guilty, but we fail to convict him (fail to reject H0).
   * That’s a Type II error.
* Statistical example:
   * H0: A new cancer drug has no effect.
   * We conclude “no effect” (fail to reject H0), even though it actually saves lives.

5. What is the difference between a Z-test and a T-test? Explain when to use
each?

* **Z-Test**
* A statistical test used to determine if there is a significant difference between sample and population means (or between two samples), when the population variance (σ²) is known and/or the sample size is large.
* Distribution used: Standard normal distribution (Z-distribution).
* When to use a Z-test:
  * Large sample size (n>30) → Central Limit Theorem ensures normality.
  * Population standard deviation (σ) is known.
  * Data are approximately normally distributed.
* Example:
  * A company claims the average weight of chips packets is 50g.
  * You take a sample of 100 packets, population σ = 5g is known.
  * Use a Z-test to check if the mean differs from 50.
**T-Test**
* A statistical test used to determine if there is a significant difference between sample and population means (or between two samples), when the population variance is unknown and the sample size is small.
* Distribution used: Student’s t-distribution (fatter tails than normal).
*When to use a T-test:
  * Small sample size (n<30).
  * Population standard deviation (σ) is unknown (must estimate it using sample standard deviation s).
  * Data are approximately normally distributed.
* Example:
  * Testing whether the mean exam score of 25 students differs from 70, but population σ is unknown.
  * Use a T-test.

7.Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
(Include your Python code and output in the code box below.)

```python
import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu_0 = 50
# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)
# Z-statistic calculation
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))
# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
# Print results
print("Sample mean:", sample_mean)
print("Sample std deviation:", sample_std)
print("Sample size:", n)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Decision at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Significant evidence that mean != 50.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that mean != 50.")
```

8. Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
(Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
mu = 50      # true mean
sigma = 5    # true standard deviation
n = 100      # sample size

data = np.random.normal(mu, sigma, n)

#  Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample std
se = sample_std / np.sqrt(n)       # standard error

# 95% confidence interval using Z-distribution
z_critical = norm.ppf(0.975)  # 97.5th percentile
ci_lower = sample_mean - z_critical * se
ci_upper = sample_mean + z_critical * se

# Print results
print("Sample mean:", sample_mean)
print("95% Confidence Interval:", (ci_lower, ci_upper))

# Step 4: Plot histogram of data
plt.hist(data, bins=15, edgecolor='black', alpha=0.7, density=True)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label=f"Lower 95% CI = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label=f"Upper 95% CI = {ci_upper:.2f}")

plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

9. Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
(Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset.
    """
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation
    z_scores = [(x - mean) / std for x in data]
    return np.array(z_scores), mean, std

# Example dataset
np.random.seed(42)
data = np.random.normal(100, 15, 200)

# Calculate Z-scores
z_scores, mean, std = calculate_z_scores(data)

# Print first 5 Z-scores
print("First 5 Z-scores:", z_scores[:5])
print("Dataset mean:", mean)
print("Dataset std deviation:", std)

# Plot histogram of Z-scores
plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7, density=True)
plt.title("Histogram of Z-scores (Standardized Data)")
plt.xlabel("Z-score")
plt.ylabel("Density")

# Add reference lines for mean and ±1, ±2 standard deviations
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (Z=0)")
plt.axvline(1, color='green', linestyle='dashed', linewidth=2, label="Z=+1")
plt.axvline(-1, color='green', linestyle='dashed', linewidth=2, label="Z=-1")
plt.axvline(2, color='blue', linestyle='dashed', linewidth=2, label="Z=+2")
plt.axvline(-2, color='blue', linestyle='dashed', linewidth=2, label="Z=-2")

plt.legend()
plt.show()
```
